```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
# read the combined txt file into dataframe
mta = pd.read_csv("/Users/monkeyjiashuo/Documents/GitHub/sf19_ds15/curriculum/project-01/Project-Benson/Data Files/combined_file.txt")
```

```python
mta.columns = mta.columns.str.replace(' ', '')
```

```python
# check the # of rows&columns and data types in dataframe
mta.info()
```

```python
# randomly sample 50 instances to get sense of dataframe
mta.sample(10, random_state = 22)
```

```python
# transfor DATE column to datatime
mta['DATE'] = pd.to_datetime(mta['DATE'], format = '%m/%d/%Y')
```

```python
# transform Date column and create a new column Day of week
mta['DAY_OF_WEEK'] = mta['DATE'].dt.strftime('%A')
```

```python
# create key column combining 'C/A','UNIT','SCP','STATION'
mta['KEY'] = mta[['C/A','UNIT','SCP','STATION']].apply(tuple, axis = 1)
```

```python
# rearrange column order
# tidious way: mta = mta[['KEY', 'C/A', 'UNIT', 'SCP', 'STATION', 'LINENAME' ,'DIVISION', 'DATE', 'TIME', 'DESC', 'ENTRIES', 'EXITS']]
cols = mta.columns.tolist();
cols = cols[-1:] + cols[:-1];
mta = mta[cols]
```

```python
mta.head()
```

```python
mta['ENTRIES_DIFF'] = mta['ENTRIES'].diff(1);
mta['EXITS_DIFF'] = mta['EXITS'].diff(1);
mta['TOTAL_TRAFFIC'] = mta['ENTRIES_DIFF'] + mta['EXITS_DIFF']
```

```python
mta.head(5)
```

```python
mta.loc[mta['ENTRIES_DIFF'] <= 0.0, 'ENTRIES_DIFF'] = np.nan;
mta.loc[mta['ENTRIES_DIFF'] > 1000000.0, 'ENTRIES_DIFF'] = np.nan;
mta.loc[mta['EXITS_DIFF'] <= 0.0, 'EXITS_DIFF'] = np.nan;
mta.loc[mta['EXITS_DIFF'] > 1000000.0, 'EXITS_DIFF'] = np.nan;
mta.loc[mta['TOTAL_TRAFFIC'] <= 0.0, 'TOTAL_TRAFFIC'] = np.nan;
mta.loc[mta['TOTAL_TRAFFIC'] > 1000000.0, 'TOTAL_TRAFFIC'] = np.nan;
```

```python
mta = mta.dropna()
```
